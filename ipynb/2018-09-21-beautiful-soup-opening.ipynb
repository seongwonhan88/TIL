```python
from bs4 import BeautifulSoup

with open('weekday.html') as f:
    html = f.read()
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
#select all anchor with class=title
for a in soup.select('a.title'):
    print(a.get_text())
```

```python
img_list = soup.select('div.thumb img')
for img in img_list:
    print(img['src'])
```
